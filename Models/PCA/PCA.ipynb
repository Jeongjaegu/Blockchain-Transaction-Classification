# PCA

```python
import numpy as np
import pandas as pd
import tensorflow as tf
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
```

```python
df = pd.read_csv("Data/user_rec.csv", sep = ",") #640 MB
df.head(5)
```

```python
df_sub = df[:100000]
```

```python
df_sub = df[-100000:]
```

```python
df_sub = tf.keras.utils.normalize(np.array(df_sub)) 
```

```python
pca = PCA(n_components=3)
pca.fit(df_sub)
```

```python
pca = PCA(n_components = 2)  # project from 8 to 2 dimensions
projected = pca.fit_transform(df_sub)
print(df_sub.data.shape)
print(projected.shape)
```

```python
plt.scatter(projected[:,0],projected[:,1], c=df_sub[:,1]) # Last 100000 users
#plt.axis([0, 6, 0, ])
plt.show()
```

```python
plt.scatter(projected[:,0],projected[:,1], c=df_sub[:,-1]) # First 100000 users
#plt.axis([0, 6, 0, ])
plt.show()
```

```python
print(pca.explained_variance_ratio_)  
```
